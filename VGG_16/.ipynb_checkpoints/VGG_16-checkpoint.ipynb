```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import torch
from torch.autograd import Variable
import torch.nn as NN
from torch.optim import Adam, SGD
```

### Prepare Dataset

```python
train = pd.read_csv('emergency_vs_non-emergency_dataset/emergency_train.csv')
train.head()
```

```python
train_img = []
for img_name in tqdm(train['image_names']):
    
    img_path = 'emergency_vs_non-emergency_dataset/images/' + str(img_name)
    
    img = imread(img_path)
    
    img = img/255
    #mode = numpy.pad i.e. If the image is smaller than the specified resize value, it'll pad it with zeros.
    #default mode if not specified is 'reflect'.
    img = resize(img, output_shape=(224,224,3), mode='constant', anti_aliasing=True)
    
    img = img.astype('float32')
    
    train_img.append(img)

train_x = np.array(train_img)
train_y = train['emergency_or_not'].values
train_x.shape
```

**Visualize Dataset**

```python
fig = plt.figure(figsize=(20,10))
fig.subplots_adjust(hspace=0, wspace=0.2)
column, row = 10,1

for i in range(0, row*column):
    
    img = None
    title = None
    
    img = train_x[i]
    title = str(train_y[i])
    
    ax = fig.add_subplot(row,column, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.xticks(())
    plt.yticks(())
plt.show()
```

**Split Dataset**

```python
#stratified sampling ensures both train and test portion have the same percentage of classes.
train_x, val_x, train_y, val_y = train_test_split(train_x, train_y, test_size=0.2, random_state= 42, stratify=train_y)
```

**Convert dataset into PyTorch format**

```python
train_x = train_x.reshape(train_x.shape[0], 3, 224,224)
train_x = torch.from_numpy(train_x)

train_y = train_y.astype(int)
train_y = torch.from_numpy(train_y)

val_x = val_x.reshape(val_x.shape[0], 3, 224,224)
val_x = torch.from_numpy(val_x)

val_y = val_y.astype(int)
val_y = torch.from_numpy(val_y)
```

### VGG-16

```python
class VGG_16(NN.Module):
    
    def _initialize_weights(self):
            
        for m in self.modules():

            if isinstance(m, NN.Conv2d):
                NN.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    NN.init.constant_(m.bias, 0)
            elif isinstance(m, NN.BatchNorm2d):
                NN.init.constant_(m.weight, 1)
                NN.init.constant_(m.bias, 0)
            elif isinstance(m, NN.Linear):
                NN.init.normal_(m.weight, 0, 0.01)
                NN.init.constant_(m.bias, 0)      
    
    def __init__(self, num_classes=2, init_weights=True):
        
        super(VGG_16, self).__init__()
        
        self.cfgs = {
            #architecture of VGG-16. M stands for max-pooling. The numbers are the output channel for each layer.
            'vgg_16':[64,64, 'M', 128,128, 'M', 256, 256, 256, 'M', 512, 512, 512, 'M', 512, 512, 512, 'M']
        }
        
        layers = []
        in_channels = 3
        
        #construct the convolutional layers based on the configuration.
        for v in self.cfgs['vgg_16']:
            
            if v == 'M': #max pooling
                layers += [NN.MaxPool2d(kernel_size=2, stride=2)]
            else:
                #convolutional layer with input and output channel as specified.
                conv2d = NN.Conv2d(in_channels, v, kernel_size=3, padding=1)
                layers += [conv2d, NN.BatchNorm2d(num_features=v), NN.ReLU(inplace=True)]
                in_channels=v #update
        
        self.features= NN.Sequential(*layers) #unpack the layers
        self.avgpool = NN.AdaptiveAvgPool2d((7,7)) #NOT SURE WHY THIS IS REQUIRED.
        
        #NN.sequential makes sure the layers are executed in order.
        ##The fully-connected layers were suppose to output 4096 channels. I reduced it to 2048 for this problem.
        self.classifier = NN.Sequential(
                            NN.Linear(512*7*7, 2048),
                            NN.ReLU(inplace=True),
                            NN.Dropout(),
                            NN.Linear(2048,2048),
                            NN.ReLU(inplace=True),
                            NN.Dropout(),
                            NN.Linear(2048, num_classes)
        )
        
        if init_weights:
            self._initialize_weights()
            
    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x
    

vgg = VGG_16()
vgg.load_state_dict(torch.load('./model_ckpt/model.pth'))
#vgg.eval() #this is required for inference. It'll set dropout and batch normalization layers to eval mode.
optimizer= Adam(vgg.parameters(), lr=1e-4)
criterion = NN.CrossEntropyLoss() #One-Hot encoding is already implemented in CEL.

if torch.cuda.is_available():
    
    vgg = vgg.cuda()
    criterion = criterion.cuda()

print(vgg)
```

**Validation**

```python
testing_loss = []
accuracy_count = 0
for i in tqdm(range(0, val_x.size()[0], batch_size)):
        
        
    batch_x, batch_y = val_x[i:i+batch_size], val_y[i:i+batch_size]

    if torch.cuda.is_available():
        batch_x, batch_y = batch_x.cuda(), batch_y.cuda()

    outputs = vgg(batch_x)
    loss = criterion(outputs, batch_y)

    testing_loss.append(loss.item())
        
        
    _, predicted = torch.max(outputs,1)
    
    prediction = predicted.cpu().detach().numpy()
    label      = batch_y.cpu().detach().numpy()
    
    correct_count = np.where(prediction==label) #returns the indices where both have matching values.
    accuracy_count += correct_count[0].shape[0] #Get the num of correct counts.
    
print("The Validation accuracy is : ", (accuracy_count/val_x.size()[0])*100)

torch.save(vgg.state_dict(), './model_ckpt/model.pth')
```
