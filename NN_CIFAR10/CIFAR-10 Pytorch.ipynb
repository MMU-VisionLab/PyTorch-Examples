```python
import numpy as np
import torch #to access utils module
import torchvision #to access dataset
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import torch.nn as nn #NN modules
import torch.nn.functional as F 
import torch.optim as optim #Torch optimization functions
```

### Dataset Preparation

```python
batch_size = 16
```

```python
#Compose consists of one or more transformation process on the raw dataset. In this example, we're transforming
#the images into Tensor and normalizes them. 
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))])

#import CIFAR10 training dataset while applying the transform created earlier.
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
#loader
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size, shuffle=True, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
def image_show(img):
    npimg = img.numpy()/2 + 0.5 #un-normalize the images back
    plt.imshow(np.transpose(npimg, (1,2,0)))
    plt.show()

dataiter = iter(trainloader)
images, labels = dataiter.next()

image_show(torchvision.utils.make_grid(images))
print(' '.join('%2s' % classes[labels[j]] for j in range(batch_size)))
```

### CNN

```python
#inherit from torch's NN module.
class Net(nn.Module):
    
    def __init__(self):
        #initialize the parent class
        super(Net, self).__init__()
        #This defines a convolutional layer. It accepts 3 channels as our image is in RGB mode and it 
        #outputs 6 channels. Strides are 1 by default and there padding=0 by default. Bias are added by default.
        self.conv1 = nn.Conv2d(in_channels=3, out_channels= 6, kernel_size=5, padding=2) 
        self.pool = nn.MaxPool2d(2,2) #A standard 2 by 2 max pooling layer
        self.conv2 = nn.Conv2d(6,16,5) #6 inp-channels, 16 out-channels and kernel_size=5
        self.fc1 = nn.Linear(6*6*16, 120) #The input is first max_pooled then convolved and max pooled again, hence 
        #ending up with 7x7 feature map.
        self.fc2 = nn.Linear(120,80)
        self.fc3 = nn.Linear(80,10) #10 as the output is of 10 classes.
        
        #WEIGHT INITIALIZATION (Not necessary)
        #options to initialize
        torch.nn.init.xavier_uniform_(self.conv1.weight)
#         self.conv2.weight.data.fill_(0.01)
#         self.conv2.bias.data.fill_(0.1) 
        
    def forward(self, x): 
        #Feed forward function, this function must be written by all classes that inherit from nn.Module
        first_conv = self.pool(F.relu(self.conv1(x)))
        sec_conv = self.pool(F.relu(self.conv2(first_conv)))
        reshape_tensor = sec_conv.view(-1, 6*6*16)
        first_fc = F.relu(self.fc1(reshape_tensor))
        sec_fc = F.relu(self.fc2(first_fc))
        third_fc = self.fc3(sec_fc) #no act func at the output node.
        return third_fc, sec_fc, first_fc #normally you'd want to only return the third_fc (output node)
    
#     def backward() -> this function
        
net = Net()
print(net) #shows information about the defined nodes
```

```python
#TESTING --can be ignored--
out1,out2,out3 = net.forward(images[0].view(1,3, 32,32))
print(out1.shape) #output node

for p in net.parameters(): #all the parameters in the network
    print(p) #As can be seen, conv2 is initialized with 0.01
```

```python
#Loss func and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)
```

```python
#training

for epoch in range(3):
    
    epoch_loss = 0.0
    
    for i, data in enumerate(trainloader, 0):
        
        inputs, labels = data
        
        #We have to initialize the gradient to zero between mini-batches since the optimizer accumulates the 
        #gradient unlike tensorflow's static graph. This accumulation property can be used in RNN.
        optimizer.zero_grad() 
        
        #get the output from the network (prediction)
        outputs, _,_ = net(inputs)
        
        loss = criterion(outputs, labels) #calculate the loss
        loss.backward() #computes the gradient for every parameter that has require_grad=True
        
        optimizer.step() #performs the optimization based on the computed gradient
#         print(optimizer.step) can be printed to show the information about the optimizer.
        
        epoch_loss += loss.item()
        
    print("Epoch %d, Loss %g"%(epoch, epoch_loss))
```

```python
#save model parameters
save_path = './cifar_net.pth'
torch.save(net.state_dict(), save_path)
```
