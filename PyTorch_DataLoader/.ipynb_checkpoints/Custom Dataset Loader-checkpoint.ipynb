# PyTorch DataLoaders

**PyTorch comes with many tools to make data loading easier and enables many optimization techniques to be used on the loading so that data loading doesn't become a bottleneck in the training procedure.**

```python
import os
import torch
import pandas as pd
from skimage import io,transform
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
```

### Sample

```python
landmarks_frame = pd.read_csv('faces/face_landmarks.csv')

n=60
img_name = landmarks_frame.iloc[n, 0]
landmarks = landmarks_frame.iloc[n, 1:]
landmarks = np.asarray(landmarks)
landmarks = landmarks.astype('float').reshape(-1,2)

print("Image name: {}".format(img_name))
print("Landmarks shape: {}".format(landmarks.shape))
print("First 4 Landmarks: {}".format(landmarks[:4]))
```

```python
def show_landmarks(image, landmarks):
    
    plt.imshow(image)
    plt.scatter(landmarks[:, 0], landmarks[:, 1], s=10, marker='.', c='r')
    plt.pause(0.001)

plt.figure()
show_landmarks(io.imread('faces/'+img_name), landmarks)
plt.show()
```

### Dataset Class by PyTorch

torch.utils.data.Dataset is an abstract class. It should be inherited and overridden with the following methods:

    a) __len__ : Returns the size of the dataset
    b) __getitem__ : Returns the i-th dataset

```python
class FaceLandmarksDataset(Dataset):
    
    def __init__(self, csv_file, root_dir, transform=None): #We will be reading the csv file in __init__
        '''
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        '''
        self.landmarks_frame = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform
    
    def __len__(self):
        
        return len(self.landmarks_frame)
    
    def __getitem__(self, idx):
        
        if torch.is_tensor(idx):
            idx = idx.tolist()
        
        img_name = os.path.join(self.root_dir, self.landmarks_frame.iloc[idx, 0])
        image = io.imread(img_name)
        landmarks = self.landmarks_frame.iloc[idx, 1:]
        landmarks = np.array([landmarks])
        landmarks = landmarks.astype('float').reshape(-1, 2)
        sample = {'image': image, 'landmarks':landmarks}
        
        if self.transform:
            sample = self.transform(sample)
        
        return sample        
```

```python
face_dataset = FaceLandmarksDataset(csv_file='faces/face_landmarks.csv',
                                   root_dir='faces/')

fig = plt.figure()

for i in range(len(face_dataset)):
    sample = face_dataset[i]

    print(i, sample['image'].shape, sample['landmarks'].shape)

    ax = plt.subplot(1, 4, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(i))
    ax.axis('off')
    show_landmarks(**sample)

    if i == 2:
        plt.show()
        break
```

**The images are not of the same size. Since we want a constant size for the input of neural networks and possibly some data augmentation, we can make use of the Transform class.**

```python
class Rescale:
    
    
    def __init__(self, output_size):
        assert isinstance(output_size, (int, tuple)) #Check if the output size is an int or tuple.
        self.output_size = output_size
    
    def __call__(self, sample):
        image, landmarks = sample['image'], sample['landmarks']
        
        h, w = image.shape[:2]
        
        #we want to maintain the aspect ratio of images if the input is an integer.
        if isinstance(self.output_size, int):
            if h>w:
                new_h, new_w = self.output_size*h/w, self.output_size
            else:
                new_h, new_w = self.output_size, self.output_size*w/h
        else:
            new_h, new_w = self.output_size
        
        new_h, new_w = int(new_h), int(new_w)
        
        img = transform.resize(image, (new_h, new_w))
        
        #we also have to resize the landmarks positions accordingly. Since the landmarks are coordinated 
        #on x axis first then y axis, we use width first then height.
        landmarks = landmarks * [new_w/w, new_h/h]
        
        return {'image':img, 'landmarks':landmarks}
    
class ToTensor:
    
    def __call__(self, sample):
        
        image, landmarks = sample['image'], sample['landmarks']
        
        image = image.transpose((2,0,1)) #since pytroch requires (channels, height, width)
        return {'image': torch.from_numpy(image),
               'landmarks': torch.from_numpy(landmarks)}
```

```python
composed = transforms.Compose([Rescale((224,224))]) #We can compose more transformations if we had implemented them
fig = plt.figure()

for i, tsfrm in enumerate([composed]):
    transformed_sample = tsfrm(sample)
    
    ax = plt.subplot(1, 3, i+1)
    plt.tight_layout()
    ax.set_title(type(tsfrm).__name__)
    show_landmarks(**transformed_sample)
```

### Iterating through the Dataset

```python
transformed_dataset = FaceLandmarksDataset(csv_file='faces/face_landmarks.csv',
                                          root_dir='faces/',
                                          transform=transforms.Compose([
                                              Rescale((224,224)),
                                              ToTensor()
                                          ]))
#This way, we're losing out on batch_size, multiprocessing and shuffling.
for i in range(len(transformed_dataset)):
    sample = transformed_dataset[i]

    print(i, sample['image'].size(), sample['landmarks'].size())

    if i == 3:
        break
```

**PyTorch DataLoader**

```python
dataloader = DataLoader(transformed_dataset, batch_size=5, shuffle=True, num_workers=5)


for i_batch, sample_batched in enumerate(dataloader):
    print(i_batch, sample_batched['image'].size(),
          sample_batched['landmarks'].size())
```
